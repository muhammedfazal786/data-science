## Univariate

## Histplot

```python
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
```

```python
n_cols = 3
n_rows = (len(num_cols) + n_cols - 1) // n_cols
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, 8))
axes = axes.flatten()
for i, col in enumerate(num_cols):
    sns.histplot(df[col], kde=True, color="#ea00ffea", ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')
for j in range(i + 1, n_rows * n_cols):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

## Feature,Key Observation
### Oldpeak : "Huge spike at 0 (~350), right-skewed tail â†’ many true zeros (no ST depression), keep but impute suspicious 0s"
### Cholesterol: "Spike at 0 (~200), then normal from 100â€“400 â†’ 0 = missing, replace with NaN & impute"
### MaxHR: "Clean bell shape (100â€“180) â†’ perfect, no action"
### Age: "Smooth, slightly right-skewed (30â€“70) â†’ ready to use"
### RestingBP: "Multi-peaks (120, 130, 140), long right tail â†’ rounded values, normal, optional cap >180"
